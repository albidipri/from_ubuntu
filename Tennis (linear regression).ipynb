```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python

df = pd.read_csv('tennis_stats.csv')
df.head()
```

```python
df.keys()
```

```python
plt.scatter(df['ReturnGamesWon'], df['Wins'])
```

```python
plt.scatter(df['ServiceGamesWon'], df['Wins'])
```

```python
plt.scatter(df['Aces'], df['Winnings'])
```

```python
plt.scatter(df['BreakPointsOpportunities'], df['Winnings'])
```

```python
X = df[['BreakPointsOpportunities']]
y = df[['Winnings']]
```

```python
X_train, X_test,y_train, y_test = train_test_split(X, y, train_size = 0.8)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
prediction = model.predict(X_test)
plt.scatter(y_test,prediction, alpha=0.4)
```

```python
features = df[['Aces']]
outcome = df[['Winnings']]
```

```python
features_train, features_test, outcome_train, outcome_test = train_test_split(features, outcome, train_size = 0.8)
```

```python
model = LinearRegression()
model.fit(features_train,outcome_train)
```

```python
prediction = model.predict(features_test)
plt.scatter(outcome_test,prediction, alpha=0.4)
```

```python
model.score(X_test,y_test)
```

```python
model.score(features_test,outcome_test)
```
